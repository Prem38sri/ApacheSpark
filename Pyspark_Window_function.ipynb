## rank , dense_rank and row_number

```
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, rank, dense_rank, col, sum
```

```
windowspec = Window.partitionBy("COUNTRY").orderBy("SALES")
windowspec_1 = Window.partitionBy("PRODUCTLINE", "COUNTRY")
```

```
df = dataset.select("SALES", "PRODUCTLINE", "COUNTRY")
df.display()
#df_rank_by_row = df.withColumn("row_number", row_number().over(windowspec))
```

```
df1 = df.withColumn("SALES", sum(col("SALES")).over(windowspec_1)).distinct()
```

```
df2 = df1.withColumn("ranking_by_row", row_number().over(windowspec))
```

```
df2.display()
```

```
df3 = df1.withColumn("ranking_by_rank", rank().over(windowspec))
```

## LAG and LEAD

```
dataset.display()
```

```
df_lag_and_lead = dataset.select("SALES", "QTR_ID", "YEAR_ID")
```

```
df_lag_and_lead.display()
```

```
from pyspark.sql.functions import lag, lead, concat, lit
windowspec_for_sales_by_qtr_year = Window.partitionBy("QTR_ID", "YEAR_ID").orderBy("YEAR_ID", "QTR_ID")
```

```
df_for_lag = df_lag_and_lead.withColumn("SALES", sum(col("SALES")).over(windowspec_for_sales_by_qtr_year)).distinct().orderBy("YEAR_ID", "QTR_ID")
```

```
df_for_lag.display()
```

```
df_for_lag_1 = df_for_lag.withColumn("PREVIOUS_SALES", lag("SALES", 4).over(Window.orderBy("YEAR_ID", "QTR_ID"))).withColumn("NEXT_YEAR", lead("SALES", 4).over(Window.orderBy("YEAR_ID", "QTR_ID")))
```

```
df_for_lag_2 = df_for_lag_1.withColumn("QOQ", ((col("SALES") - col("PREVIOUS_SALES")) / col("PREVIOUS_SALES")) * 100)
```

```
df_for_lag_3 = df_for_lag_2.filter(col("QOQ").isNotNull()).select("QTR_ID", "YEAR_ID", "QOQ").withColumn("YEAR_AND_QTR", concat(col("YEAR_ID"), lit("0"), col("QTR_ID"))).select("YEAR_AND_QTR", "QOQ")
```

```
df_for_lag_3.display()
```

```
df = dataset.select("SALES", "QTR_ID", "YEAR_ID")
```

```
df1 = df.groupBy("QTR_ID", "YEAR_ID").sum("SALES").orderBy("QTR_ID", "YEAR_ID")
```
