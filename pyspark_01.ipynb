```
data = spark.sparkContext.textFile("/FileStore/tables/sample.txt")
```

```
data1 = data.flatMap(lambda x: x.split(" "))
```

```
data2 = data1.map(lambda x: x.title())
```

```
data2.collect()
```

```
def filter_out_2(partition):
  for element in partition:
    st = element.split(" ")
    for s in st:
      s = s.title()
      yield s

```

```
filtered_lists = data.mapPartitions(filter_out_2)
```

```
type(filtered_lists)
```

```
filtered_lists.collect()
```
